###Random Walk en una dimensión

En este primer código de Python, vamos a realizar el ejercicio uno desarrollado en excel, veremos un paseo aleatorio en una dimensión:

```python
import random
import matplotlib.pyplot as plt
random.seed()
Accion = 1000     # inicializamos el valor actual de Tesla
n = 31             # indicamos el numero de dias de cotizacion que queremos saber
factor = 1.02       # como su variación media es del 1%, el factor sera de 1.01
p=0.5               # la probalidad que suba la acccion un 1% es del 50%
Precio = [Accion] # inicializamos la serie de precios, comenzamos con la primera cotización
for t in range(1,n):
    if random.random() < p:
        Accion *= factor
    else:
        Accion /= factor
    Precio.append(Accion)
plt.plot(range(n),Precio)
plt.show()               #Mostramos el grafico de la cotizacion de Tesla durante los 30 dias
Varianza= (Accion/999) - 1
print(f"La cotizacion de tesla ha variado un {Varianza:.2%} ")    #dividimos el precio final de tesla entre el inicial 

                                                                  #para saber en que porcentaje ha variado su cotizacion
```

En este segundo código, realizaremos el ejercicio desarrollado en excel, donde vemos un random walk realizado en dos dimensiones:


```python
# Ahora siguiendo el jemeplo anterior podemos ver como se mueve una accion en dos dimensiones de manera aleatoria
# Se parte del punto (0,0) del plano cartesiano  
# Cada paso que se da es de longitud r  
# El ángulo con el que se da cada paso es aleatorio  
import random  
import math  
import matplotlib.pyplot as plt  
random.seed()  
n=31 # número de pasos  
r=1 # longitud del paso  
x=y=0 # punto de partida (x,y)=(0,0)  
listax=[x]  
listay=[y]  
for i in range(n):  
  angulo=random.random()*math.pi*2 # en radianes  
  x+=r*math.cos(angulo)  
  y+=r*math.sin(angulo)  
  listax.append(x)  
  listay.append(y)  
plt.plot(listax,listay)  
plt.show()
```

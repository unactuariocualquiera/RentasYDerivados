###Random Walk en una dimensión

En este primer código de Python, vamos a realizar el ejercicio uno desarrollado en excel, veremos un paseo aleatorio en una dimensión:

```python
import random
import matplotlib.pyplot as plt
random.seed()
Accion = 1000     #Valor Inicial
n = 31             #Tiempo en días que vamos a generar la aleatoriedad
factor = 1.02       #Variación
p=0.5               #Probabilida de aumento o disminución (50%)
Precio = [Accion] 
for t in range(1,n):
    if random.random() < p:
        Accion *= factor
    else:
        Accion /= factor
    Precio.append(Accion)

#Mostramos el gráfico de los valores aleatorios generados
plt.plot(range(n),Precio)
plt.show()               
Varianza= (Accion/999) - 1
print(f"La variación de la serie aleatoria es {Varianza:.2%} ")  #Conocemos la variación final
```

En este segundo código, realizaremos el ejercicio desarrollado en excel, donde vemos un random walk realizado en dos dimensiones:


```python
#Partimos del punto (0,0) teniendo en cuenta el plano cartesiano
#En esta ocasión el paso que da no es un número natural, será r

import random  
import math  
import matplotlib.pyplot as plt  
random.seed()  
n=31 #Cantidad de pasos generados
r=1  
x=y=0  #Punto de partida comentado anteriormente
listax=[x]  
listay=[y]  
for i in range(n):  
  angulo=random.random()*math.pi*2 
  x+=r*math.cos(angulo)  
  y+=r*math.sin(angulo)  
  listax.append(x)  
  listay.append(y)  
plt.plot(listax,listay)  
plt.show()
```
